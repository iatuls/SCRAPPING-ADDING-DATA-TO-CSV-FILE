```python
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
import logging
import pandas as pd
```

```python
url = "https://rentrayage.com/collections/sale"
r = requests.get(url)
soup = bs(r.text, 'html.parser')
bigbox=soup.find_all('h3',{'class':'product-title'})
price=soup.find_all('span',class_='product-price')
```

```python

product=[]
for i in bigbox:
    product.append(i.text)
    # product_name = i.text
```

```python

prices=[]
for i in price:
        # product_price = i.text
        prices.append(i.text)
        # product_name = i.text
```

```python
Product_Name = pd.DataFrame(product, columns=['Product Name'])
Product_Price = pd.DataFrame(prices, columns=['Product Price'])
```
